```python
import pandas as pd

import sports
import academic
import general
import tech
```

```python
gen_word = []

gen_len = (len(general.general))

for i in range(0, gen_len) :
    gen_word.append('general')
    
```

```python
tech_word = []

tech_len = (len(tech.tech))

for i in range(0, tech_len) :
    tech_word.append('tech')
```

```python
sports_word = []

sports_len = (len(sports.sports))

for i in range(0, sports_len) :
    sports_word.append('sports')
```

```python
aca_word = []

aca_len = (len(academic.academic))

for i in range(0, aca_len) :
    aca_word.append('academic')
```

```python
master_values = (general.general + tech.tech + sports.sports + academic.academic )
master_values
```

```python
master_categories = (gen_word + tech_word + sports_word + aca_word)
master_categories
```

```python
data = pd.DataFrame([general.general, tech.tech, sports.sports, academic.academic], index=['General', 'Tech', 'Sports', 'Academic']).T
data
```

```python
data = pd.DataFrame([master_values, master_categories], index = ['Name', 'Category']).T
data
```

```python
compression_opts = dict(method='zip',
                        archive_name='raw_dataset.csv')  
data.to_csv('data.zip', index=False,
          compression=compression_opts)
```
